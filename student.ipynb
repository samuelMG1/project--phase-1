## Data Analysis

### Analysis of accidents distribution over time

```python
df['Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Date'].dt.year
accidents_per_year = df['Year'].value_counts().sort_index()
accidents_per_year
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Plotting the graph showing trends of accidents

```python
sns.lineplot(x=accidents_per_year.index, y=accidents_per_year.values)
plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

From the visualization above, we can note that:

Before 1980, there were no aeroplanes. The year 1980 recorded the highest number of accidents. As the years progress, the number of accidents is reducing due to improvements in skilled manpower and technology, though more research needs to be done.

```python
Aircrafts = df.groupby('Aircraft.Category')['Make'].count()
Aircrafts   
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Aircraft.Category', order=df['Aircraft.Category'].value_counts().index, color= 'green')
plt.title('Distribution of Accidents by Aircraft Category') # 
plt.xlabel('Aircraft Category')
plt.ylabel('Number of Accidents') 
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.yscale('log')  # Use a logarithmic scale for better visualization of the data
plt.show()
```

### Accidents over time by make

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=df[df['Make'].isin(top_makes)], x='Make', order=top_makes)
plt.title('Top 10 Makes with Most Accidents')
plt.xlabel('Make')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

### Distribution of Total Fatal, serious and minor injuries

```python
phase_of_flight_data = df.groupby('Broad.phase.of.flight').agg(
    total_incidents=('Event.Id', 'count'),
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum')
).reset_index()
phase_of_flight_data
```

```python
plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_incidents'], color='blue')
plt.title('Total Incidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Incidents')
plt.xticks(rotation=45)
plt.grid(True)
plt.subplot(2, 1, 2)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_fatal_injuries'], color='red', alpha=0.5, label='Fatal Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_serious_injuries'], color='pink', alpha=0.5, label='Serious Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_minor_injuries'], color='black', alpha=0.5, label='Minor Injuries')
plt.title('Total Injuries by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Injuries')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

plotting: Two subplots are created:

The first subplot shows the total number of incidents by phase of flight. The second subplot shows the total number of fatal, serious, and minor injuries by phase of flight, differentiated by color (hue). This visualization will help you understand the distribution of incidents and injuries across different phases of flight.
